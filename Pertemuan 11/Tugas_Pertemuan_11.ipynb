```python
import datetime
import string
import random

# Template dict buku
buku_template = {
    "judul": "judul",
    "penulis": "penulis",
    "tahun_terbit": 0,
    "tanggal_terakhir_pinjam": datetime.datetime(1111, 1, 11),
    "status": "status"
}

perpustakaan = {}

while True:
    print(f"{"SELAMAT DATANG":^20}")
    print(f"{"SISTEM PERPUSTAKAAN":^20}")
    print("-" * 20)

    buku = dict.fromkeys(buku_template.keys())
    buku["judul"] = input("Judul Buku: ")
    buku["penulis"] = input("Penulis: ")
    buku["tahun_terbit"] = int(input("Tahun Terbit: "))
    buku["status"] = input("Status (TERSEDIA / TIDAK): ").upper()

    # Input tanggal terakhir pinjam (jika belum pernah, input 0 untuk tahun)
    TAHUN_PINJAM = int(input("Tahun Terakhir Pinjam (YYYY, atau 0 jika belum pernah): "))
    if TAHUN_PINJAM == 0:
        buku["tanggal_terakhir_pinjam"] = None
    else:
        BULAN_PINJAM = int(input("Bulan Terakhir Pinjam (1-12): "))
        TANGGAL_PINJAM = int(input("Tanggal Terakhir Pinjam (1-31): "))
        buku["tanggal_terakhir_pinjam"] = datetime.datetime(TAHUN_PINJAM, BULAN_PINJAM, TANGGAL_PINJAM)

    KEY = "".join((random.choice(string.ascii_uppercase) for i in range(6)))
    perpustakaan.update({KEY: buku})

    print(f"\n{"KEY":<6} {"Judul":<20} {"Penulis":<15} {"Tahun":<6} {"Tgl Pinjam":<12} {"Status":<10}")
    print("-" * 80)

    for id_buku in perpustakaan:
        KEY = id_buku
        JUDUL = perpustakaan[KEY]["judul"]
        PENULIS = perpustakaan[KEY]["penulis"]
        TAHUN = perpustakaan[KEY]["tahun_terbit"]
        TGL_PINJAM = perpustakaan[KEY]["tanggal_terakhir_pinjam"].strftime("%x") if perpustakaan[KEY]["tanggal_terakhir_pinjam"] else "Belum"
        STATUS = perpustakaan[KEY]["status"]
        print(f"{KEY:<6} {JUDUL:<20} {PENULIS:<15} {TAHUN:<6} {TGL_PINJAM:<12} {STATUS:<10}")

    print("\n")
    is_done = input("Mau tambah lagi bro (y/n)? ")
    if is_done == "n":
        break

print("\nProgram berakhir")
```
