Jawaban nomor 1

```python
buah = ['apel','mangga','jeruk','anggur','pisang']

# a. Tambahkan semangka di akhir list
buah.append('semangka')
print(f'list buah setelah semangka ditambah = {buah}\n')

# b. Sisipkan 'durian' di antara 'jeruk' dan 'anggur'
buah.insert(2,'durian')
print(f'list buah setelah durian disisipkan = {buah}\n')

# c. Hapus 'mangga' dari list
buah.remove('mangga')
print(f'list buah setelah mangga dihapus = {buah}\n')

# d. Ubah 'pisang' menjadi 'nanas'
buah[4] = 'nanas'
print(f'list buah setelah pisang diubah menjadi nanas = {buah}\n')

# e. Tampilkan 3 buah pertama
buah = ['apel','mangga','jeruk','anggur','pisang']

buah.append('semangka')
buah.insert(3, 'durian')
buah.remove('mangga')
buah[buah.index('pisang')] = 'nanas'

print('list akhir = ', buah)
print('\n3 buah pertama = ', buah[:3])
```

Jawaban nomor 2

```python
angka = [45, 12, 78, 23, 56, 89, 34]

# a. Urutkan list secara ascending (terkecil ke terbesar)
angka.sort()
print(f'angka setelah diurutkan secara ascending = {angka}\n')

# b. Urutkan list secara descending (terbesar ke terkecil)
angka.reverse()
print(f"angka setelah diurutkan secara descending = {angka}")
```
